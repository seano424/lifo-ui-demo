## 2. Perfect CSV Processing Demo

Let's start with a clean, well-formatted CSV file to see the processor in action.

```python
def demo_clean_csv_processing():
    """Demo: Processing perfect CSV data"""
    
    print("🟢 DEMO: Processing Clean CSV Data")
    print("=" * 50)
    
    # Load perfect CSV
    perfect_csv = "../data/clean_data/perfect_inventory.csv"
    
    # Show original data first
    df_original = pd.read_csv(perfect_csv)
    print(f"📊 Original data shape: {df_original.shape}")
    print(f"📋 Original columns: {list(df_original.columns)}")
    print("\n📝 First few rows:")
    display(df_original.head())
    
    # Process with UnifiedCSVProcessor
    processor = UnifiedCSVProcessor(store_id="demo-store-001")
    
    print("\n🔄 Processing with UnifiedCSVProcessor...")
    result = processor.process_csv_file(perfect_csv)
    
    # Show results
    print(f"\n✅ Processing Status: {result['status']}")
    print(f"📈 Success Rate: {result['success_rate']:.1f}%")
    print(f"📊 Processed Count: {result['processed_count']}/{result['metadata']['total_rows']}")
    print(f"⚠️ Warnings: {len(result['warnings'])}")
    print(f"❌ Errors: {len(result['errors'])}")
    
    # Show warnings if any
    if result['warnings']:
        print("\n⚠️ Warnings:")
        for warning in result['warnings']:
            print(f"   • {warning}")
    
    # Show processed data sample
    if result['data']:
        print("\n📋 Processed Data Sample:")
        processed_df = pd.DataFrame(result['data'])
        display(processed_df.head())
    
    return result

# Run the demo
clean_result = demo_clean_csv_processing()
```

## 3. Messy Data Processing Demo

Now let's see how the processor handles messy, real-world data with various issues.

```python
def demo_messy_csv_processing():
    """Demo: Processing messy CSV data with cleaning"""
    
    print("\n🟡 DEMO: Processing Messy CSV Data")
    print("=" * 50)
    
    messy_files = [
        "../data/messy_data/mixed_formats.csv",
        "../data/messy_data/missing_columns.csv",
        "../data/messy_data/duplicate_skus.csv",
        "../data/messy_data/invalid_data.csv"
    ]
    
    results = {}
    
    for file_path in messy_files:
        print(f"\n📁 Processing: {os.path.basename(file_path)}")
        print("-" * 40)
        
        # Show original data issues
        try:
            df_original = pd.read_csv(file_path)
            print(f"📊 Original shape: {df_original.shape}")
            print(f"📋 Original columns: {list(df_original.columns)}")
            
            # Show sample of problematic data
            print("\n📝 Sample data:")
            display(df_original.head(3))
            
            # Process with error handling
            processor = UnifiedCSVProcessor(store_id="demo-store-001")
            result = processor.process_csv_file(file_path)
            
            results[file_path] = result
            
            # Show processing results
            print(f"\n✅ Status: {result['status']}")
            print(f"📈 Success Rate: {result['success_rate']:.1f}%")
            print(f"📊 Processed: {result['processed_count']}/{result['metadata']['total_rows']}")
            
            if result['warnings']:
                print(f"⚠️ Warnings ({len(result['warnings'])}):") 
                for warning in result['warnings'][:3]:  # Show first 3
                    print(f"   • {warning}")
                if len(result['warnings']) > 3:
                    print(f"   ... and {len(result['warnings']) - 3} more")
            
            if result['errors']:
                print(f"❌ Errors ({len(result['errors'])}):") 
                for error in result['errors'][:3]:  # Show first 3
                    print(f"   • {error}")
                if len(result['errors']) > 3:
                    print(f"   ... and {len(result['errors']) - 3} more")
                    
        except Exception as e:
            print(f"❌ Failed to process {file_path}: {e}")
            results[file_path] = {'status': 'failed', 'error': str(e)}
    
    return results

# Run the demo
messy_results = demo_messy_csv_processing()
```

## 4. Edge Cases and Security Testing

Let's test the processor's security features and edge case handling.

```python
def demo_edge_cases_and_security():
    """Demo: Testing edge cases and security features"""
    
    print("\n🔒 DEMO: Edge Cases and Security Testing")
    print("=" * 50)
    
    edge_files = [
        "../data/edge_cases/security_test.csv",
        "../data/edge_cases/empty_file.csv",
        "../data/edge_cases/single_row.csv"
    ]
    
    results = {}
    
    for file_path in edge_files:
        print(f"\n🔍 Testing: {os.path.basename(file_path)}")
        print("-" * 40)
        
        try:
            # Show what we're testing
            df_original = pd.read_csv(file_path)
            print(f"📊 File shape: {df_original.shape}")
            
            if len(df_original) > 0:
                print("\n📝 Content sample:")
                display(df_original.head())
            
            # Process with security checks
            processor = UnifiedCSVProcessor(store_id="demo-store-001")
            result = processor.process_csv_file(file_path)
            
            results[file_path] = result
            
            # Show security results
            print(f"\n🛡️ Security Status: {result['status']}")
            print(f"📈 Success Rate: {result['success_rate']:.1f}%")
            
            if result['warnings']:
                print(f"⚠️ Security Warnings:")
                for warning in result['warnings']:
                    print(f"   • {warning}")
            
            if result['errors']:
                print(f"🚨 Security Errors:")
                for error in result['errors']:
                    print(f"   • {error}")
                    
        except Exception as e:
            print(f"❌ Processing failed: {e}")
            results[file_path] = {'status': 'failed', 'error': str(e)}
    
    return results

# Run the demo
edge_results = demo_edge_cases_and_security()
```

## 5. Data Processing Visualization

Let's create visualizations to show the processing results across all test files.

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set up plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

def visualize_processing_results(clean_result, messy_results, edge_results):
    """Create comprehensive visualizations of processing results"""
    
    print("\n📊 VISUALIZATION: Processing Results Analysis")
    print("=" * 50)
    
    # Combine all results
    all_results = {
        'perfect_inventory.csv': clean_result,
        **{os.path.basename(k): v for k, v in messy_results.items()},
        **{os.path.basename(k): v for k, v in edge_results.items()}
    }
    
    # Extract metrics
    files = []
    success_rates = []
    warning_counts = []
    error_counts = []
    processed_counts = []
    
    for file_name, result in all_results.items():
        if result and result.get('status') != 'failed':
            files.append(file_name)
            success_rates.append(result.get('success_rate', 0))
            warning_counts.append(len(result.get('warnings', [])))
            error_counts.append(len(result.get('errors', [])))
            processed_counts.append(result.get('processed_count', 0))
    
    # Create figure with subplots
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('LIFO.AI CSV Processing Results Dashboard', fontsize=16, fontweight='bold')
    
    # Success rates bar chart
    colors = ['green' if rate >= 80 else 'orange' if rate >= 50 else 'red' for rate in success_rates]
    bars1 = axes[0, 0].bar(range(len(files)), success_rates, color=colors, alpha=0.7)
    axes[0, 0].set_title('CSV Processing Success Rates', fontweight='bold')
    axes[0, 0].set_ylabel('Success Rate (%)')
    axes[0, 0].set_xticks(range(len(files)))
    axes[0, 0].set_xticklabels(files, rotation=45, ha='right')
    axes[0, 0].set_ylim(0, 100)
    
    # Add value labels on bars
    for bar, rate in zip(bars1, success_rates):
        height = bar.get_height()
        axes[0, 0].text(bar.get_x() + bar.get_width()/2., height + 1,
                       f'{rate:.1f}%', ha='center', va='bottom', fontweight='bold')
    
    # Warning and error counts
    x_pos = np.arange(len(files))
    width = 0.35
    bars2 = axes[0, 1].bar(x_pos - width/2, warning_counts, width, label='Warnings', color='orange', alpha=0.7)
    bars3 = axes[0, 1].bar(x_pos + width/2, error_counts, width, label='Errors', color='red', alpha=0.7)
    axes[0, 1].set_title('Issues Found During Processing', fontweight='bold')
    axes[0, 1].set_ylabel('Count')
    axes[0, 1].set_xticks(x_pos)
    axes[0, 1].set_xticklabels(files, rotation=45, ha='right')
    axes[0, 1].legend()
    
    # Items processed
    bars4 = axes[1, 0].bar(range(len(files)), processed_counts, color='blue', alpha=0.7)
    axes[1, 0].set_title('Items Successfully Processed', fontweight='bold')
    axes[1, 0].set_ylabel('Number of Items')
    axes[1, 0].set_xticks(range(len(files)))
    axes[1, 0].set_xticklabels(files, rotation=45, ha='right')
    
    # Add value labels
    for bar, count in zip(bars4, processed_counts):
        height = bar.get_height()
        axes[1, 0].text(bar.get_x() + bar.get_width()/2., height + 0.1,
                       f'{count}', ha='center', va='bottom', fontweight='bold')
    
    # Processing summary
    total_files = len(files)
    avg_success = np.mean(success_rates) if success_rates else 0
    total_warnings = sum(warning_counts)
    total_errors = sum(error_counts)
    total_processed = sum(processed_counts)
    
    summary_text = f"""Processing Summary:
    
📁 Files Processed: {total_files}
📈 Average Success Rate: {avg_success:.1f}%
📊 Total Items Processed: {total_processed}
⚠️ Total Warnings: {total_warnings}
❌ Total Errors: {total_errors}

🏆 Best Performer: {files[success_rates.index(max(success_rates))] if success_rates else 'N/A'}
🔧 Most Challenging: {files[success_rates.index(min(success_rates))] if success_rates else 'N/A'}

Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
"""
    
    axes[1, 1].text(0.05, 0.95, summary_text, transform=axes[1, 1].transAxes, 
                   fontsize=11, verticalalignment='top', fontfamily='monospace',
                   bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
    axes[1, 1].set_title('Processing Summary', fontweight='bold')
    axes[1, 1].axis('off')
    
    plt.tight_layout()
    plt.savefig('../outputs/visualizations/csv_processing_results.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return {
        'total_files': total_files,
        'avg_success_rate': avg_success,
        'total_processed': total_processed,
        'total_warnings': total_warnings,
        'total_errors': total_errors
    }

# Create visualizations
viz_summary = visualize_processing_results(clean_result, messy_results, edge_results)
print("\n✅ Visualization saved to: ../outputs/visualizations/csv_processing_results.png")
```

## 6. Data Quality Analysis

Let's analyze the data quality patterns we discovered during processing.

```python
def analyze_data_quality_patterns():
    """Analyze common data quality issues found"""
    
    print("\n🔍 DATA QUALITY ANALYSIS")
    print("=" * 50)
    
    # Collect all warnings and errors
    all_warnings = []
    all_errors = []
    
    # From clean result
    all_warnings.extend(clean_result.get('warnings', []))
    all_errors.extend(clean_result.get('errors', []))
    
    # From messy results
    for result in messy_results.values():
        if result and result.get('status') != 'failed':
            all_warnings.extend(result.get('warnings', []))
            all_errors.extend(result.get('errors', []))
    
    # From edge results
    for result in edge_results.values():
        if result and result.get('status') != 'failed':
            all_warnings.extend(result.get('warnings', []))
            all_errors.extend(result.get('errors', []))
    
    # Analyze warning patterns
    warning_patterns = {
        'Date Issues': [w for w in all_warnings if 'date' in w.lower()],
        'Category Issues': [w for w in all_warnings if 'category' in w.lower()],
        'Price Issues': [w for w in all_warnings if 'price' in w.lower()],
        'Duplicate Issues': [w for w in all_warnings if 'duplicate' in w.lower()],
        'Security Issues': [w for w in all_warnings if any(x in w.lower() for x in ['dangerous', 'security', 'formula'])],
        'Other Issues': [w for w in all_warnings if not any(x in w.lower() for x in ['date', 'category', 'price', 'duplicate', 'dangerous', 'security', 'formula'])]
    }
    
    # Analyze error patterns
    error_patterns = {
        'Required Field Errors': [e for e in all_errors if 'required' in e.lower()],
        'Format Errors': [e for e in all_errors if 'format' in e.lower()],
        'Validation Errors': [e for e in all_errors if 'invalid' in e.lower()],
        'Other Errors': [e for e in all_errors if not any(x in e.lower() for x in ['required', 'format', 'invalid'])]
    }
    
    # Display analysis
    print("📊 WARNING PATTERN ANALYSIS:")
    print("-" * 30)
    for pattern, warnings in warning_patterns.items():
        if warnings:
            print(f"⚠️ {pattern}: {len(warnings)} occurrences")
            for warning in warnings[:2]:  # Show first 2 examples
                print(f"   • {warning}")
            if len(warnings) > 2:
                print(f"   ... and {len(warnings) - 2} more")
            print()
    
    print("\n📊 ERROR PATTERN ANALYSIS:")
    print("-" * 30)
    for pattern, errors in error_patterns.items():
        if errors:
            print(f"❌ {pattern}: {len(errors)} occurrences")
            for error in errors[:2]:  # Show first 2 examples
                print(f"   • {error}")
            if len(errors) > 2:
                print(f"   ... and {len(errors) - 2} more")
            print()
    
    # Data quality recommendations
    print("\n💡 DATA QUALITY RECOMMENDATIONS:")
    print("-" * 35)
    
    recommendations = []
    
    if warning_patterns['Date Issues']:
        recommendations.append("📅 Standardize date formats (YYYY-MM-DD recommended)")
    
    if warning_patterns['Category Issues']:
        recommendations.append("🏷️ Use standard category names from LIFO.AI taxonomy")
    
    if warning_patterns['Price Issues']:
        recommendations.append("💰 Remove currency symbols and use decimal numbers")
    
    if warning_patterns['Duplicate Issues']:
        recommendations.append("🔄 Ensure SKU uniqueness across all products")
    
    if warning_patterns['Security Issues']:
        recommendations.append("🔒 Avoid formula-like content in CSV cells")
    
    if error_patterns['Required Field Errors']:
        recommendations.append("📋 Ensure all required fields (SKU, product name, category, quantity, expiry date) are present")
    
    if error_patterns['Format Errors']:
        recommendations.append("🔧 Validate data formats before upload")
    
    for i, rec in enumerate(recommendations, 1):
        print(f"{i}. {rec}")
    
    if not recommendations:
        print("🎉 No major data quality issues found! Your data is well-formatted.")
    
    return {
        'warning_patterns': warning_patterns,
        'error_patterns': error_patterns,
        'recommendations': recommendations
    }

# Run data quality analysis
quality_analysis = analyze_data_quality_patterns()
```

## 7. Processing Performance Summary

Let's summarize the overall performance of our CSV processing system.

```python
def generate_processing_summary():
    """Generate a comprehensive processing summary"""
    
    print("\n📋 COMPREHENSIVE PROCESSING SUMMARY")
    print("=" * 50)
    
    # Calculate overall statistics
    total_files_processed = 1 + len(messy_results) + len(edge_results)
    successful_files = 0
    total_items = 0
    successful_items = 0
    
    # Count successful files and items
    if clean_result.get('status') == 'success':
        successful_files += 1
        total_items += clean_result.get('metadata', {}).get('total_rows', 0)
        successful_items += clean_result.get('processed_count', 0)
    
    for result in messy_results.values():
        if result and result.get('status') == 'success':
            successful_files += 1
            total_items += result.get('metadata', {}).get('total_rows', 0)
            successful_items += result.get('processed_count', 0)
    
    for result in edge_results.values():
        if result and result.get('status') == 'success':
            successful_files += 1
            total_items += result.get('metadata', {}).get('total_rows', 0)
            successful_items += result.get('processed_count', 0)
    
    # Calculate success rates
    file_success_rate = (successful_files / total_files_processed) * 100
    item_success_rate = (successful_items / total_items) * 100 if total_items > 0 else 0
    
    print(f"📊 PROCESSING STATISTICS:")
    print(f"   Files Processed: {total_files_processed}")
    print(f"   Files Successful: {successful_files}")
    print(f"   File Success Rate: {file_success_rate:.1f}%")
    print(f"   Total Items: {total_items}")
    print(f"   Items Processed: {successful_items}")
    print(f"   Item Success Rate: {item_success_rate:.1f}%")
    
    print(f"\n🎯 KEY ACHIEVEMENTS:")
    achievements = [
        "✅ Successfully processed clean, well-formatted data",
        "✅ Handled messy data with intelligent error recovery",
        "✅ Detected and prevented security vulnerabilities",
        "✅ Provided comprehensive error reporting and warnings",
        "✅ Normalized inconsistent data formats",
        "✅ Validated data integrity and business rules"
    ]
    
    for achievement in achievements:
        print(f"   {achievement}")
    
    print(f"\n🔧 PROCESSOR CAPABILITIES DEMONSTRATED:")
    capabilities = [
        "🛡️ Security validation and formula injection protection",
        "🔄 Intelligent column mapping and normalization",
        "📅 Multi-format date parsing and validation",
        "💰 Price validation with currency symbol handling",
        "🏷️ Category mapping and standardization",
        "⚠️ Comprehensive warning and error reporting",
        "📊 Data quality analysis and recommendations"
    ]
    
    for capability in capabilities:
        print(f"   {capability}")
    
    print(f"\n🎉 CONCLUSION:")
    print(f"   The LIFO.AI CSV Processor successfully demonstrated its ability to:")
    print(f"   • Handle diverse data formats and quality levels")
    print(f"   • Provide robust security and validation features")
    print(f"   • Deliver actionable insights for data quality improvement")
    print(f"   • Maintain high processing success rates even with problematic data")
    
    return {
        'total_files': total_files_processed,
        'successful_files': successful_files,
        'file_success_rate': file_success_rate,
        'total_items': total_items,
        'successful_items': successful_items,
        'item_success_rate': item_success_rate
    }

# Generate final summary
processing_summary = generate_processing_summary()

print(f"\n🚀 Demo completed successfully! Ready for next steps.")
print(f"📝 Next: Try the API Integration Demo (02_API_Integration_Demo.ipynb)")
print(f"🤖 Or: Explore the Scoring Algorithm Demo (03_Scoring_Algorithm_Demo.ipynb)")
```
